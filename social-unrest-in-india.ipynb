````python
from urllib import request
import json
import geopandas as gpd
url = "https://gist.githubusercontent.com/jbrobst/56c13bbbf9d97d187fea01ca62ea5112/raw/e388c4cae20aa53cb5090210a42ebb9b765c0a36/india_states.geojson"
with request.urlopen(url) as f:
    geo_india = json.load(f)
    
geo_df = gpd.read_file(url)
geo_df["lon"] = geo_df["geometry"].centroid.x
geo_df["lat"] = geo_df["geometry"].centroid.y
````

````python
statewise_event_counts = indian_riots_df.groupby('admin1').size().reset_index(name="count")
statewise_event_counts.replace("Jammu and Kashmir","Jammu & Kashmir", inplace=True)

statewise_event_counts_dict = statewise_event_counts.set_index('admin1').to_dict()['count']
geo_df['event_count'] = geo_df['ST_NM'].apply(lambda state_name : statewise_event_counts_dict.get(state_name, 0))
````

````python
fig = go.Figure(data=go.Choropleth(
    geojson="https://gist.githubusercontent.com/jbrobst/56c13bbbf9d97d187fea01ca62ea5112/raw/e388c4cae20aa53cb5090210a42ebb9b765c0a36/india_states.geojson",
    featureidkey='properties.ST_NM',
    locationmode='geojson-id',
    locations=statewise_event_counts['admin1'],
    z=statewise_event_counts['count'],
    text=statewise_event_counts['admin1'],

    autocolorscale=False,
    colorscale='oranges',
    marker_line_color='peachpuff',
    colorbar=dict(
        title={'text': "Number of Events"},

        thickness=15,
        len=0.35
    )
))

fig.update_geos(
    visible=False,
    projection=dict(
        type='conic conformal',
        parallels=[12.472944444, 35.172805555556],
        rotation={'lat': 24, 'lon': 80}
    ),
    lonaxis={'range': [68, 98]},
    lataxis={'range': [6, 38]}
)
fig.add_trace(go.Scattergeo(
    lon=geo_df['lon'],
    lat=geo_df['lat'],
    mode='text',
    text=['{}<br>{}'.format(k,v) for k,v in zip(geo_df['ST_NM'].str.title(), geo_df['event_count'])],
    textfont={
        'color': 'black',
        'size': 9
    },
    name='',
))

fig.update_layout(
    title=dict(
        text="India - Region wise Riot Occurences",
        xanchor='center',
        x=0.5,
        yref='paper',
        yanchor='bottom',
        y=1,
        pad={'b': 10},
        
    ),
    dragmode=False,
    margin={'r': 0, 't': 30, 'l': 0, 'b': 0},
    height=700
)

fig.show()
````

* The number of protests/riots in Punjab has been significant, but the number of fatalities is relatively lower than in Jammu & Kashmir. 
* This suggests that the events in J & K are more violent compared to those in other states. 
* Similarly, Chattisgarh has a lower number of events but higher fatalities, indicating that the events that do occur tend to be more violent. 
* Therefore, it is important to consider the nature of conflicts rather than simply the frequency of events as the determining factor for fatalities. <br>

### **Question 8: Make the wordcloud using `notes` column of the dataset.**

### **Wordcloud**


> A **wordcloud is a visual representation** of textual data that highlights the most frequently occurring words in a body of text. The words are usually displayed in **varying sizes**, with **larger fonts** representing words that **occur more frequently**, while **smaller fonts represent less frequent words**. This type of visualization is useful in quickly identifying the most common themes, topics or keywords within a large body of text, making it easier to grasp the overall meaning or focus of the text at a glance. <br> Hence, this will be used to display the overall trend that the events had followed.





````python
# Seperate the lines and making an individual word.
words = ' '.join((word for word in indian_riots_df.notes.dropna().str.replace(' ', ' ')))
wc = WordCloud(collocation_threshold=int(1e6), width=1000, height=400, background_color='white').generate(words)
plt.figure(figsize=(16,8))
plt.axis("off")
plt.grid(False)
plt.imshow(wc);
````



> Various words are highlighted in the wordcloud above. This shows the frequency of how much a particular word is repeated. The more frequently a word is used, the more significancfe it has in the data. We can see the **Protest** as the biggest word which means it is used the most often. Other words such as **size, report, police, Punjab, government** and others can be seen widely.
